```python
import numpy 
import pandas as pd 
import os
import matplotlib.pyplot as plt
```

```python
train_path = "https://raw.githubusercontent.com/dbloxham1/kaggle_projects/main/bank_churn/train.csv"
train = pd.read_csv(train_path, 
                    sep = ",")
retained = train[train['Exited'] == 0]
churned = train[train['Exited'] == 1]
```

```python
train.head()
```

```python
# Plotting the histogram
plt.hist(train.loc[:,'CreditScore'], bins=10, color='blue', edgecolor='black')
plt.hist(churned.loc[:,'CreditScore'], bins=10, color='red', edgecolor='black')

# Adding a vertical line for the median
median_value = train.loc[:,'CreditScore'].median()
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label='Median')
median_value2 = churned.loc[:,'CreditScore'].median()
plt.axvline(median_value2, color='black', linestyle='dashed', linewidth=2, label='Churned Median')

# Adding labels and title
plt.xlabel('Credit Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Credit Scores')

# Display the histogram
plt.show()
```

```python
# Count the occurrences of each gender
gender_counts = train.loc[:,'Gender'].value_counts()

# Plotting the bar chart
plt.bar(gender_counts.index, gender_counts, color=['blue', 'pink'])

# Display the histogram
plt.show()
```
